```python
import zipfile
import pandas as pd
import io
from pathlib import Path
```

```python
"""
Lê arquivos Parquet de dentro de arquivos ZIP em uma pasta,
concatena todos e salva como um arquivo Parquet único.
"""
pasta_origem = r"E:\SPTrans\Pedido 097465 - dados de bilhetagem\arquivos PARQUET"
caminho_pasta = Path(pasta_origem)
lista_dataframes = []

caminho_destino = "outputs/Dados_Tarifa_Zero_2023_2025.parquet"
```

```python
# 1. Iterar sobre todos os arquivos .zip na pasta
for arquivo_zip in caminho_pasta.glob('*.zip'):
    print(f"Lendo ZIP: {arquivo_zip.name}...")
    
    # 2. Abrir o arquivo ZIP
    with zipfile.ZipFile(arquivo_zip, 'r') as z:
        # 3. Iterar sobre os arquivos internos do ZIP
        for arquivo_interno in z.namelist():
            if arquivo_interno.endswith('.parquet'):
                print(f"  -> Extraindo: {arquivo_interno}")
                
                # 4. Ler o arquivo diretamente para a memória em formato de bytes
                dados_em_bytes = z.read(arquivo_interno)
                
                # 5. Converter os bytes para um DataFrame do pandas usando io.BytesIO
                df = pd.read_parquet(io.BytesIO(dados_em_bytes), engine='fastparquet')
                lista_dataframes.append(df)
```

```python
print("\nConcatenando todos os dados...")
# 7. Concatenar todos os DataFrames da lista em um único DataFrame
df_consolidado = pd.concat(lista_dataframes, ignore_index=True)
```

```python
print(f"Salvando o arquivo consolidado em: {caminho_destino}...")
# 8. Salvar o resultado no formato Parquet
df_consolidado.to_parquet(caminho_destino, engine='fastparquet', index=False)

print(f"Sucesso! Tabela salva com {len(df_consolidado)} linhas.")
```
